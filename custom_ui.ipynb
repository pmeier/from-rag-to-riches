```python
from ragna import Rag, source_storages, assistants

chat = Rag().chat(
    documents=[
        "documents/ford-10k.pdf",
        "documents/gm-10k.pdf",
        "documents/ragna.txt",
    ],
    source_storage=source_storages.Chroma,
    assistant=assistants.Gpt35Turbo16k,
)
```

```python
await chat.prepare()
```

```python
async def callback(contents: str, user: str, instance: pn.chat.ChatInterface):
    if user == "Philip Meier":
        instance.stream(
            {
                "user": assistants.Gpt35Turbo16k.display_name(),
                "avatar": pn.chat.message.GPT_3_LOGO,
                "object": await chat.answer(contents),
            }
        )
```

```python
chat_interface = pn.chat.ChatInterface(
    callback=callback,
    callback_user="Ragna",
    user="Philip Meier",
    avatar="images/philip_meier.jpg",
    show_clear=False,
    show_undo=False,
)
```

```python
chat_interface.send(
    "How can I help you with the documents?",
    user="Ragna",
    avatar="images/ragna_logo.svg",
    respond=False,
)

chat_interface.servable()
```
